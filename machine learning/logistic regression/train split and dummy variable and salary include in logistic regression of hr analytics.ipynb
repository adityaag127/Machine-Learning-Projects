## As we can see that salary is in string that have three sepration high,medium,low so we can convert them in dummy variables

```python
dummies=pd.get_dummies(df["salary"],prefix="salary")
```

```python
dummies.head()
```

```python
merged=pd.concat([df,dummies],axis="columns")
```

```python
merged.head()
```

```python
final_merged=merged.drop(["salary"],axis="columns")
```

```python
final_merged.head()
```

```python
x=final_merged[["satisfaction_level","average_montly_hours","promotion_last_5years","salary_high","salary_low","salary_medium"]]
y=final_merged[["left"]]
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lm=LogisticRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
lm.predict(x_test)
```

```python
lm.score(x_test,y_test)
```
