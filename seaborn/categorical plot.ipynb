```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Categorical plot --> use to plot categorical data (if we have categorical data in x or y axis we use categorical data)
### catplot()
### boxplot()
### stripplot()
### swarmplot()

```python
titanic=sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
#catplot()-->like scatter plot
```

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.catplot(x="day",y="total_bill",data=tips)## catplot is used to plot the graph on categorical data
## Earlier we used numeric data to represt scatter plot so we use scatterplot or relplot
```

```python
sns.catplot(x="total_bill",y="day",data=tips)
```

```python
sns.catplot(x="day",y="total_bill",data=tips,jitter=False)## represent in a single line
```

```python
## swarm plot is like tree
```

```python
sns.catplot(x="day",y="total_bill",data=tips,kind="swarm")## kind="swarm" is like a tree
```

```python
sns.catplot(x="day",y="total_bill",data=tips,kind="swarm",hue="sex")
```

```python
sns.catplot(x="smoker",y="tip",data=tips)
```

```python
sns.catplot(x="smoker",y="tip",data=tips,order=["No","Yes"])## to reverse the order
```

```python
## If we want statics in the plot itself we use box plot
```

```python
sns.catplot(x="day",y="total_bill",kind="box",data=tips)## As we can see that it give statics and we can get what it represent form internet
```

```python
sns.catplot(x="day",y="total_bill",kind="box",data=tips,hue="sex")
```

```python
sns.catplot(x="day",y="total_bill",kind="box",data=tips,hue="sex",dodge=False)
```

```python
## Boxen Plot gives more information then box plot
```

```python
diamonds=sns.load_dataset("diamonds")
```

```python
diamonds.head()
```

```python
sns.catplot(x="color",y="price",kind="boxen",data=diamonds)
```

```python
sns.catplot(x="color",y="price",kind="boxen",data=diamonds.sort_values("color"))## As we can see that both plot are sorted by colors(based on x-axis)
```

```python
diamonds.sort_values("color")
```

```python
sns.catplot(x="day",y="total_bill",kind="boxen",data=tips,dodge=False)
```

```python
## violin plot
```

```python
sns.catplot(x="total_bill",y="day",kind="violin",data=tips)
```

```python
sns.catplot(x="total_bill",y="day",kind="violin",data=tips,hue="time")
```

```python
sns.catplot(x="total_bill",y="day",kind="violin",data=tips,hue="sex")
```

```python
sns.catplot(x="total_bill",y="day",kind="violin",data=tips,hue="sex",split=True)
```

```python
## draw swarm plot inside violin plot
```

```python
g=sns.catplot(x="day",y="total_bill",kind="violin",data=tips)
sns.swarmplot(x="day",y="total_bill",color="k",size=3,data=tips,ax=g.ax)
```

## Ploting Categorical data vs Categorical data

```python
titanic.head()
```

```python
sns.catplot(x="sex",y="survived",hue="class",kind="bar",data=titanic)## By this we can use bar plot
```

```python
sns.catplot(x="deck",kind="count",palette="ch:0.25",data=titanic)## If We want count of numbers based on category
```

```python
sns.catplot(x="sex",kind="count",palette="ch:0.25",data=titanic)## As we see male travellers are more than female
```

```python
sns.catplot(x="sex",kind="count",palette="ch:0.25",data=titanic,hue="class")
```

```python
sns.catplot(x="sex",y="survived",hue="class",kind="point",data=titanic)## Point plot give the last point of the bar chart
```
