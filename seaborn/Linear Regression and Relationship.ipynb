## Linear Regression and Relationship
### regplot
### lmplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.regplot(x="total_bill",y="tip",data=tips)## By regression plot we get the best fitted line with scatter plot or best plot it can give depend on data
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips)## regplot and lmplot are same
```

```python
sns.lmplot(x="size",y="tip",data=tips)
```

```python
sns.regplot(x="size",y="tip",data=tips)
```

```python
sns.lmplot(x="size",y="tip",data=tips,x_jitter=0.05)
```

```python
sns.lmplot(x="size",y="tip",data=tips,x_estimator=np.mean)
```
