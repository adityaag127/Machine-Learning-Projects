```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
def sinplot(flip = 1):
    x=np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x,np.sin(x+i*0.5)*(7-i)*flip)
```

```python
sinplot()
```

```python
sns.set_style("whitegrid")
sinplot()
```

```python
sns.set_style("darkgrid")
sinplot()
```

```python
sns.set_style("ticks")
sinplot()
```

```python
sns.set_style("ticks")
sinplot()
sns.despine()
```

```python
sns.set_style("ticks")
sinplot()
sns.despine(left=True,bottom=True)
```

```python
sns.set_style("dark")
sinplot()
sns.despine(left=True,bottom=True)
```
