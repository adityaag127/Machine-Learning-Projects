### kdeplot()
### jointplot()

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
iris=sns.load_dataset("iris")
```

```python
iris.head()
```

```python
sns.distplot(iris["sepal_length"])## distplot is used to see how data is distributed(plot only on one variable)
```

```python
sns.distplot(iris["sepal_length"],kde=False)
```

```python
sns.distplot(iris["sepal_length"],hist=False,rug=False)
```

```python
sns.kdeplot(iris["sepal_length"],shade=True)## kde plot is also used in distributed data(in one variable)
```

```python
sns.kdeplot(iris["sepal_length"],shade=True,bw=0.2)
```

```python
sns.kdeplot(iris["sepal_length"],shade=True,cut=0)
```

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

```python
x=tips["total_bill"]
y=tips["tip"]
```

```python
sns.jointplot(x=x,y=y)## joint plot is like a scatter plot with some extra information(having 3 plots in it)
```

```python
sns.jointplot(x=x,y=y,kind="hex")
```

```python
sns.jointplot(x=x,y=y,hue="smoker",data=tips)
```

```python
sns.jointplot(x=x,y=y,kind="kde",color="m")
```

## pairplot() : To plot multiple pairwise bivariate distributions in a dataset, you can use the pairplot() function. This shows the relationship for (n, 2) combination of variable in a DataFrame as a matrix of plots and the diagonal plots are the univariate plots

```python
iris.head()
```

```python
sns.pairplot(iris)## As we see that diagonal elements have same axis(or we can say that only one axis) thats why the are in bar or hist or dist or kde
```

```python
g=sns.PairGrid(iris)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.scatterplot)
```
