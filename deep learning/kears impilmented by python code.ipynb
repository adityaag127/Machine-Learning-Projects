```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
df=pd.read_csv("data.csv")
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df[["age","affordability"]],df.have_insurance,test_size=0.2)
```

```python
x_train
```

```python
#Scaling
x_train_scaled=x_train.copy()
x_train_scaled['age']=x_train_scaled['age']/100
x_test_scaled=x_test.copy()
x_test_scaled['age']=x_test_scaled['age']/100
```

```python
model=keras.Sequential([
    keras.layers.Dense(1,input_shape=(2,),activation='sigmoid',kernel_initializer='ones',bias_initializer='zeros')
])
model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])
model.fit(x_train_scaled,y_train,epochs=5000)
```

```python
model.evaluate(x_test_scaled,y_test)
```

```python
def sigmoid(x):
    import math
    return 1/(1+math.exp(-x))
sigmoid(18)
```

```python
def prediction_function(age,affordability):
    weigthed_sum=coef[0]*age+coef[1]*affordability+intercept
    return sigmoid(weigthed_sum)
```

```python
def log_loss(y_true,y_predicted):
    epsilon=1e-15
    y_predicted_new=[max(i,epsilon) for i in y_predicted]
    y_predicted_new=[min(i,1-epsilon) for i in y_predicted]
    y_predicted_new=np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
def sigmoid_numpy(x):
    return 1/(1+np.exp(-x))
sigmoid_numpy(np.array([12,0,1]))
```

```python
def gradient_descent(age,affordability,y_true,epochs):
        w1=w2=1
        bias=0
        rate=0.5
        n=len(age)
        for i in range(epochs):
            weigthed_sum=w1*age+w2*affordability+bias
            y_predicted=sigmoid_numpy(weigthed_sum)
            loss=log_loss(y_true,y_predicted)
            w1d=(1/n)*np.dot(np.transpose(age),(y_predicted-y_true))
            w2d=(1/n)*np.dot(np.transpose(affordability),(y_predicted-y_true))
            bias_d=np.mean(y_predicted-y_true)
            w1=w1-rate*w1d
            w2=w2-rate*w2d
            bias=bias-rate*bias_d
            print(f'Epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}')
        return w1,w2,bias
```
